## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Real World Examples

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
# Basic graph

plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)

# Brackets work too, good for headings with spaces
plt.plot(gas['Year'], gas['South Korea'])

plt.show()
```

```python
# Fix up the graph to be clearer

plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time in USD', fontdict={'fontweight': 'bold', 'fontsize': 22})

plt.plot(gas.Year, gas.USA, 'b.-', label='Unites States')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')

# Brackets work too, good for headings with spaces
plt.plot(gas['Year'], gas['South Korea'], 'g.-', label='South Korea')

# x axis every 3 years
plt.xticks(gas.Year[::3])

plt.xlabel = 'Year'
plt.ylabel = 'USD'

plt.legend()

plt.show()
```

```python
# Can also iterate over csv

plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time in USD')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)

        plt.xticks(gas.Year[::3])

plt.xlabel = 'Year'
plt.ylabel = 'USD'

plt.legend()

plt.show()
```

## FIFA Data

```python
fifa = pd.read_csv('fifa_data.csv')

fifa
```

### Histograms

```python

plt.xlabel = 'Skill Level'
plt.ylabel = 'Number of Players'


bins = [40, 50, 60, 70, 80, 90, 100]

plt.hist(fifa.Overall, bins = bins)

plt.xticks = bins

plt.show()
```

```python
# Just show first 5 rows
fifa.head(5)
```

### Pie Chart

```python
# pandas method
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]

right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors = colors, autopct = '%.2f')

plt.show()
```
