## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Real World Examples

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
# Basic graph

plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)

# Brackets work too, good for headings with spaces
plt.plot(gas['Year'], gas['South Korea'])

plt.show()
```

```python
# Fix up the graph to be clearer

plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time in USD')

plt.plot(gas.Year, gas.USA, 'b.-', label='Unites States')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')

# Brackets work too, good for headings with spaces
plt.plot(gas['Year'], gas['South Korea'], 'g.-', label='South Korea')

# x axis every 3 years
plt.xticks(gas.Year[::3])

plt.xlabel = 'Year'
plt.ylabel = 'USD'

plt.legend()

plt.show()
```

```python
# Can also iterate over csv

plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time in USD')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)

        plt.xticks(gas.Year[::3])

plt.xlabel = 'Year'
plt.ylabel = 'USD'

plt.legend()

plt.show()
```
